```python
#defining levels of 3-point shooting

def three_pt_team (t_pt):
    if t_pt >= 38: 
        return 'High'
    elif t_pt >= 34:
        return 'Medium'
    elif t_pt >= 30:
        return 'Low'
    elif t_pt < 30:
        return 'Very low'
    else:
        return t_pt
```

```python
basket_csv['3P_O'].apply(three_pt_team)
```

```python
basket_csv['3pt shooting']=basket_csv['3P_O'].apply(three_pt_team)
```

```python
#создаём переменные, которые генерализируют данные об атаке и защите команд
```

```python
#ranging teams by their defense

def defensive_rating (DF_EF):
    if DF_EF <= 90: 
        return 'Best defense'
    elif DF_EF <= 97:
        return 'Good defense'
    elif DF_EF <= 103:
        return 'Regular defense'
    elif DF_EF <= 110:
        return 'Bad defense'
    elif DF_EF > 110:
        return 'Worst defense'
    else:
        return DF_EF
```

```python
#ranging teams by their offense

def offensive_rating (OF_EF):
    if OF_EF >= 116: 
        return 'Best offense'
    elif OF_EF >= 110:
        return 'Good offense'
    elif OF_EF >= 100:
        return 'Regular offense'
    elif OF_EF >= 90:
        return 'Bad offense'
    elif OF_EF < 90:
        return 'Worst offense'
    else:
        return OF_EF
```

```python
basket_csv['ADJOE'].max()
```

```python
basket_csv['ADJOE'].min()
```

```python
basket_csv['ADJOE'].apply(offensive_rating)
```

```python
basket_csv['ADJDE'].apply(defensive_rating)
```

```python
basket_csv['defensive rating']=basket_csv['ADJDE'].apply(defensive_rating)
```

```python
basket_csv['offensive rating']=basket_csv['ADJOE'].apply(offensive_rating)
```

```python
basket_csv['defensive rating']=pd.Categorical(basket_csv['defensive rating'])
```
